Para cargar el archivo `.csv` a una tabla pasando su ubicación a través de la variable `dataset_path`:

```python
dataset_path = 'C:/Users/Usuario/Downloads/'
dataset_file = 'competencia_01_crudo.csv'
```

```python
%%sql
create or replace table competencia_01_crudo as
select
    *
from read_csv_auto("{{dataset_path + dataset_file}}")
```

Hagamos unas queries básicas para comprobar que todo esta funcionando bien.

```python
%sql select * from competencia_01_crudo limit 5
```

```python
%%sql
-- Selecciona las filas con numero_de_cliente repetido en la tabla competencia_01_crudo
SELECT *
FROM competencia_01_crudo
WHERE numero_de_cliente IN (
    SELECT numero_de_cliente
    FROM competencia_01_crudo
    GROUP BY numero_de_cliente
    HAVING COUNT(*) > 1
)
```

```python
%%sql
select
    foto_mes
    , count(*) as cantidad -- cuenta cuantos casos hay en cada foto_mes
                           -- y lo guarda en un campo llamado cantidad
from competencia_01_crudo
group by foto_mes
```

Perfecto, ahora cree una nueva tabla con la variable adicional que se le pide.

```python
%%sql
CREATE OR REPLACE TABLE competencia_01 AS
SELECT 
    numero_de_cliente,
    MAX(CASE WHEN foto_mes = 202101 THEN 1 ELSE 0 END) AS foto_mes_202101,
    MAX(CASE WHEN foto_mes = 202102 THEN 1 ELSE 0 END) AS foto_mes_202102,
    MAX(CASE WHEN foto_mes = 202103 THEN 1 ELSE 0 END) AS foto_mes_202103,
    MAX(CASE WHEN foto_mes = 202104 THEN 1 ELSE 0 END) AS foto_mes_202104,
    MAX(CASE WHEN foto_mes = 202105 THEN 1 ELSE 0 END) AS foto_mes_202105,
    MAX(CASE WHEN foto_mes = 202106 THEN 1 ELSE 0 END) AS foto_mes_202106
FROM 
    competencia_01_crudo
GROUP BY
    numero_de_cliente
```

```python
%%sql
CREATE OR REPLACE TABLE competencia_02 AS
SELECT
    numero_de_cliente,
    CASE 
        WHEN foto_mes_202103 = 1 AND foto_mes_202104 = 0 THEN 'BAJA+1'
        WHEN foto_mes_202104 = 1 AND foto_mes_202105 = 0 THEN 'BAJA+2'
        ELSE 'CONTINUA'
    END AS clase
FROM competencia_01;
```

```python
%%sql
SELECT *
FROM competencia_02
WHERE clase = 'CONTINUA'
```
